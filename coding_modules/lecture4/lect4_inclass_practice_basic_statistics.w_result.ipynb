```python
import numpy as np
stats_array = np.full([5,4],0)
```

```python
for i_var,var in enumerate(df.columns):
    stats_array[0,i_var] = np.mean(df[var].values)
    stats_array[1,i_var] = np.percentile(df[var].values,25)
    stats_array[2,i_var] = np.median(df[var].values)
    stats_array[3,i_var] = np.percentile(df[var].values,75)
    stats_array[4,i_var] = np.std(df[var].values)
    
```

```python
df_stats = pd.DataFrame(stats_array,index = ['mean','25pctl','median','75pctl','SD'],
                        columns=df.columns.values)
```
